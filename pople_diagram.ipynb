```python
import matplotlib.pyplot as plt

# Create the figure and axis
fig, ax = plt.subplots(figsize=(6, 5))

# Set the limits and remove ticks for a conceptual look
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xticks([])
ax.set_yticks([])

# Add the main axis arrows
gap = 0.9  # Gap from the origin for better visibility
ax.annotate('', xy=(9.5, 0.5), xytext=(0.5 + gap, 0.5),
            arrowprops=dict(arrowstyle='->', lw=1.5))
ax.annotate('', xy=(0.5, 9.5), xytext=(0.5, 0.5 + gap),
            arrowprops=dict(arrowstyle='->', lw=1.5))

# Axis Labels
ax.set_xlabel('One-electron basis set size', fontsize=15, labelpad=10)
ax.set_ylabel('Expansion of $N$-electron models', fontsize=15, labelpad=10)

# The "Exact" point
ax.text(9, 9.5, 'Exact', fontsize=14, fontweight='bold', ha='center')

# The main diagonal arrow (The path to truth!)
ax.annotate('', xy=(8, 7.5), xytext=(3, 3),
            arrowprops=dict(arrowstyle='->', color='chocolate', lw=2.5))

# Adding the descriptive text along the axes
ax.text(0.7, 7, 'Toward the exact\nsolution in a\ngiven basis', 
        fontsize=14, rotation=0, va='center')

ax.text(7.5, 0.8, 'Toward CBS limit for a\ngiven $N$-electron model', 
        fontsize=14, ha='center')

# Optional: Add specific method labels if you want to make it more detailed
# ax.text(0.6, 1, 'HF', fontsize=10)
# ax.text(0.6, 9, 'FCI', fontsize=10)
# ax.text(9, 0.6, 'CBS', fontsize=10)

# Clean up spines
for spine in ax.spines.values():
    spine.set_visible(True)

plt.tight_layout()
plt.savefig('pople_diagram.pdf')
plt.show()
```
