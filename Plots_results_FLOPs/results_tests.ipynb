```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

colors = sns.color_palette("tab10")
```

# Helium

```python
N = 4488
N_occ = 1
ls = [5, 10, 30, 50, 100, 200]

FLOPS_enh = np.array([
    7584140884,
    16214360168,
    33980005948,
    59366572640,
    124025559208,
    247987358624
])

FLOPs_block = np.array([
    7121700760,
    21000525040,
    61179466320,
    177117745200,
    1739179828000,
    20432261921600
])
exact_diago = 20 * (N)**3   
```

```python
plt.figure(figsize=(8,6))
x = np.array(ls)
plt.axhline(y=exact_diago, color=colors[3], linestyle='--', label='Exact Diagonalization')
plt.plot(x, FLOPs_block, marker='d', label='Block-Davidson', color=colors[0])
plt.plot(x, FLOPS_enh, marker='o', label='Enhanced JD', color=colors[1])
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of computed eigenvalues')
plt.ylabel('FLOPs ')
xticks = np.array([10, 50, 100, 200]) * N_occ
# Create labels: one line for nev, second line for multiplier
xtick_labels = [f"{mult}$N_{{occ}}$" for nev, mult in zip(xticks, [10, 50, 100, 200])]
plt.xticks(xticks, xtick_labels)
plt.title(r'FLOPs Comparison for helium (${N_{occ}} = 1$)')
plt.legend()
plt.grid(True)
sns.despine()
# plt.tight_layout()
plt.savefig('FLOPS_comparison_Helium.pdf')
plt.show()
```

# hBN

```python
N = 6863
N_occ = 36
ls_hBN = [5, 10, 50, 100]

FLOPS_enh = np.array([
    744132114090,
    1645872428214,
    8566113213584,
    19755278727710
])

FLOPs_block = np.array([
    1339595275860,
    4109973644520,
    16046163737328,
    11293129091928
])
exact_diago = 20 * (N)**3   
```

```python
plt.figure(figsize=(8,6))
x = np.array(ls_hBN) * N_occ
plt.axhline(y=exact_diago, color=colors[3], linestyle='--', label='Exact Diagonalization')
plt.plot(x, FLOPs_block, marker='d', label='Block-Davidson', color=colors[0])
plt.plot(x, FLOPS_enh, marker='o', label='Enhanced JD', color=colors[1])
plt.ylim(6e11, 1.1e14)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of computed eigenvalues')
plt.ylabel('FLOPs ')
xticks = np.array([5, 10, 50, 100]) * N_occ
# Create labels: one line for nev, second line for multiplier
xtick_labels = [f"{mult}$N_{{occ}}$" for nev, mult in zip(xticks, [5, 10, 50, 100])]
plt.xticks(xticks, xtick_labels)
plt.title(r'FLOPs Comparison for hexagonal boron nitride (${N_{occ}} = 36$)')
plt.legend()
plt.grid(True)
sns.despine()
# plt.tight_layout()
plt.savefig('FLOPS_comparison_hBN.pdf')
plt.show()
```
