```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
from matplotlib.ticker import LogLocator

# sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

colors = sns.color_palette("tab10")
```

```python
factor = np.array([0.5, 0.7, 1, 1.5, 2, 2.5])
N = np.array([100, 200])

formaldehyde_50 = np.array([688256345911013, 710641783500873, 63019505592433, 16242097824685, 13368249830289, 18965919642037])
formaldehyde_100 = np.array([907127583220174,  1007127583220174, 1307127583220174, 22348217655421, 24934551468586, 25214942741175])
formaldehyde_200 = np.array([2051833170341421, 4051833170341421, 5051833170341421, 33487479213871, 37366877152400, 45201057403130])

exact_formaldehyde =  20 * 27643**3
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
x_ticks_labels = [f"{x:.1f}" for x in factor]

x_ticks1 = factor * N[0]
axes[0].plot(x_ticks1, formaldehyde_100 / 1e12, marker='^', label=r'$N=100 \times N_{\text{occ}}$', color=colors[0])
# axes[0].plot(x_ticks1, [exact_formaldehyde / 1e12]*len(x_ticks1), linestyle='--', label='Exact', color='black')
axes[0].set_yscale('log')
axes[0].set_xlabel(r'Naux / $n_{\text{ev}}$')
axes[0].set_title(r'$n_{\text{ev}}=100 \times N_{\text{occ}}$')
axes[0].set_xticks(x_ticks1, labels=x_ticks_labels) 
axes[0].set_ylim(7e0, 2e4)
axes[0].grid()

x_ticks2 = factor * N[1]
axes[1].plot(x_ticks2, formaldehyde_200 / 1e12, marker='x', label=r'$N=200 \times N_{\text{occ}}$', color=colors[2])
# axes[1].plot(x_ticks2, [exact_formaldehyde / 1e12]*len(x_ticks2), linestyle='--', label='Exact', color='black')
axes[1].set_yscale('log')
axes[1].set_xlabel(r'Naux / $n_{\text{ev}}$')
axes[1].set_title(r'$n_{\text{ev}}=200 \times N_{\text{occ}}$')
axes[1].set_xticks(x_ticks2, labels=x_ticks_labels)     
axes[1].set_ylim(7e0, 2e4)
axes[1].grid()

axes[0].set_ylabel(r'Flops ($\times 1e12$)')
# sns.despine()
plt.tight_layout()
plt.savefig('Naux_analysis.pdf')
plt.show()
```
