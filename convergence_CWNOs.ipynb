```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("notebook", font_scale=1.5)
# Data for approx MP2 NOs (Purple)
x_mp2 = [4, 10, 15, 20, 30, 40, 60, 80, 100]
y_mp2 = [2.3e-2, 2.1e-2, 5.5e-3, 1.3e-3, 1.3e-3, 5.2e-4, 6.5e-5, 1.3e-6, 1.2e-5]

# Data for Coulomb weighted NOs (Green)
x_cwno = [4, 10, 15, 20, 30, 40, 60, 80, 100, 150, 200]
y_cwno = [3.5e-2, 9.0e-3, 8.0e-3, 3.5e-2, 4.5e-3, 6.2e-3, 2.8e-3, 6.5e-4, 1.0e-3, 3.3e-4, 1.7e-4]

colors = sns.color_palette("tab10")
colors
```

```python
plt.figure(figsize=(10, 6))

# Plotting the lines
plt.plot(x_mp2, y_mp2, marker='^', color=colors[0], linewidth=2, label='Approx MP2 NOs')
plt.plot(x_cwno, y_cwno, marker='^', color=colors[3], linewidth=2, label='Coulomb weighted NOs')

# Formatting
plt.yscale('log')
plt.xlabel('Number of virtual orbitals per occupied orbital')
plt.ylabel('Error in MP2 correlation [eV]')
plt.title('Interaction energy of water dimer')
plt.grid()
plt.legend()
plt.tight_layout()
plt.savefig('convergence_CWNOs.pdf')
plt.show()
```
